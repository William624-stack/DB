```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
y_train
```

Criando e Treinando o Modelo

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

Avaliação Modelo

```python
# Printandoo a intercepção
print(lm.intercept_)
```

```python
X2 = sm.add_constant(X_train)
est = sm.OLS(y_train, X2)
est2 = est.fit()
print(est2.summary())
# Valor-P for menor do 5% => há relação entre X e y
# Rodamos o modelo ML completo e retiramos os atributos com valores-p superiores a 5%
# Como o sinal do coeficiente positivo, X se correclaciona com y positivamente
# R2 é acima de 50%, o ajuste dos atributos é significante
# 92% dos preços das casas são influenciados pelos artibutos considerados
```

```python
r2=lm.score(X_test, y_test)
print(r2)
```
