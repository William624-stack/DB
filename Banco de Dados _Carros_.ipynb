# Iterando com DataFrames

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.set_option.html

```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/daniel-usp/LIT/master/db.csv' , sep = ';')
```

```python
dataset.head()
```

```python
dataset['km_m√©dia'] = (dataset.Quilometragem) + 10000
dataset.head(5)
```

```python
dataset['km_m√©dia'] = (dataset.Quilometragem/(2019-dataset.Ano)).fillna(0)
dataset
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset['Quilometragem'].isna()
```

```python
len(dataset['Quilometragem'].isna())
```

```python
filtro = dataset['Quilometragem'].isna()
dataset[filtro]
```

```python
dataset.fillna(0, inplace = True)
dataset
```

```python
dataset_ordem_decrescente_valor = dataset.sort_values(by = "Valor", ascending= False)
dataset_ordem_decrescente_valor
```

```python
dataset.groupby(by = ["Motor"])["Valor"].describe().sort_values(by = "mean", ascending = False)
```

```python
dataset.groupby(by = ["Ano"])["Quilometragem"].describe().sort_values(by = "mean", ascending = False)
```
